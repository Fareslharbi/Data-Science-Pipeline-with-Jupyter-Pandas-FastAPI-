```python
import os 
import pandas as pd 
```

```python
cwd = os.getcwd() # dirname(abspath(__file__))
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'data')
CACHE_DIR = os.path.join(BASE_DIR, 'cache')
working_file = os.path.join(CACHE_DIR, 'movies-box-office-dataset.csv')
output_file = os.path.join(CACHE_DIR, 'movies-box-office-dataset-cleaned.csv')
```

```python
df = pd.read_csv(working_file)
```

```python
df.head(n=1200)
```

```python
df['Rank'] = -1
```

```python
df.head()
```

```python
df['domestic %'] = df['%']
df['foregin %'] = df['%.1']
df.head()
```

```python
#df.drop(columns=['%', '%.1'])
df.drop(['%', '%.1'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Worldwide'] = df['Worldwide'].replace("$", "").replace(",", "")
df.head()
```

```python
to_clean_col = ['Worldwide', 'Domestic', 'Foreign']
def currency_str_to_int(current_val):
    currency_val = current_val.replace("$", "").replace(",", "")
    try:
        currency_val = int(currency_val)
    except:
        currency_val = 0
    return currency_val
def clean_col(row):
    """
    row = Pandas Series
    dataframe = collection of pandas series
    """
    # print(row)
    for col in to_clean_col:
        current_val = row[col]
        row[col] = currency_str_to_int(current_val)
    return row
df_cleaned = df.apply(clean_col, axis=1)
df_cleaned.head()
```

```python
df_cleaned.dtypes
```

```python
df.dtypes
```

```python
df_cleaned.sort_values(by=['Worldwide'], inplace=True, ascending=False)
df_cleaned.head()
```

```python
df_cleaned.reset_index(inplace=True, drop=True)
df_cleaned.head()
```

```python
df_cleaned['Rank'] = df_cleaned.index + 1
df_cleaned.head()
```

```python
df_cleaned['domestic %'] =  df_cleaned['Domestic'] / df_cleaned['Worldwide'] 
df_cleaned['foregin %'] =  df_cleaned['Foreign'] / df_cleaned['Worldwide'] 
```

```python
df_cleaned.head()
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.to_csv(output_file, index=False)
```
